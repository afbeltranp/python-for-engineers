### **¿Qué es una Variable en Python?**
Una variable en Python actúa como una caja de almacenamiento que contiene datos. Es una etiqueta que puedes asociar a un valor, lo que te permite recordarlo y modificarlo en cualquier punto de tu programa. Por ejemplo, si creas una variable `age` y le asignas el valor `25` usando `age = 25`, acabas de almacenar el número `25` en una "caja" etiquetada como `age`.

---

### **Entender Qué Tipos de Variables Simples Existen en Python**
Python soporta varios tipos básicos de variables para manejar diferentes tipos de información. Aquí hay algunos:

- **Enteros (`int`)**: Números enteros sin punto decimal, como `7` o `-3`.
- **Flotantes (`float`)**: Números reales con punto decimal, como `3.14` o `-0.001`.
- **Cadenas (`str`)**: Secuencias de caracteres para texto, encerradas entre comillas como `"¡Hola, mundo!"`.
- **Booleanos (`bool`)**: Representan valores de verdad, ya sea `True` (Verdadero) o `False` (Falso).

---

### **Usar Variables para Almacenar Información**
Las variables ayudan a hacer tu código flexible y legible. Al usarlas para almacenar información, puedes actualizar los datos sin cambiar el resto del programa. Por ejemplo, si tienes una variable `temperature = 75`, y el clima cambia, puedes actualizarla así: `temperature = 68`. Esto mantiene tu código dinámico, reflejando el estado más reciente sin necesidad de recrear la variable.

```python
import ipywidgets as widgets
from IPython.display import display
import operator

def check_answer(variable_name, expected, comparison = operator.eq):
    output = widgets.Output()
    button = widgets.Button(description="Check answer")
    def _inner_check(button):
        with output:
            if comparison(globals()[variable_name], expected):
                output.outputs = [{'name': 'stdout', 'text': 'Correct!', 'output_type': 'stream'}]
            else:
                output.outputs = [{'name': 'stdout', 'text': 'Incorrect!', 'output_type': 'stream'}]
    button.on_click(_inner_check)
    display(button, output)
```
