### **What is a Variable in Python?**
A variable in Python acts like a storage box that holds data. It's a label you can attach to a value, which allows you to recall and modify it at any point in your program. For instance, if you create a variable `age` and assign it the value `25` using `age = 25`, you've just stored the number `25` in a "box" labeled `age`.

---

### **Understand What Simple Variable Types Exist in Python**
Python supports several basic types of variables to handle different kinds of information. Here are a few:

- **Integers (`int`)**: Whole numbers without decimal points, such as `7` or `-3`.
- **Floats (`float`)**: Real numbers with decimal points, like `3.14` or `-0.001`.
- **Strings (`str`)**: Sequences of characters for text, enclosed in quotes like `"Hello, world!"`.
- **Booleans (`bool`)**: Represents truth values, either `True` or `False`.

---

### **Use Variables to Store Information**
Variables help make your code flexible and readable. By using them to store information, you can update the data without changing the rest of the program. For example, if you have a variable `temperature = 75`, and the weather changes, you can update it like this: `temperature = 68`. This keeps your code dynamic, reflecting the latest state without needing to recreate the variable.

```python
import ipywidgets as widgets
from IPython.display import display
import operator

def check_answer(variable_name, expected, comparison = operator.eq):
    output = widgets.Output()
    button = widgets.Button(description="Check answer")
    def _inner_check(button):
        with output:
            if comparison(globals()[variable_name], expected):
                output.outputs = [{'name': 'stdout', 'text': 'Correct!', 'output_type': 'stream'}]
            else:
                output.outputs = [{'name': 'stdout', 'text': 'Incorrect!', 'output_type': 'stream'}]
    button.on_click(_inner_check)
    display(button, output)
```
