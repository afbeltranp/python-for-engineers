---

### **Importancia de las Extensiones de Archivo**
Las extensiones de archivo son críticas porque le indican al sistema operativo cómo manejar un archivo al identificar su tipo y el programa asociado. Aquí hay algunas extensiones de archivo comunes:

- **.txt** — Archivo de texto plano
- **.docx** — Documento de Microsoft Word
- **.pdf** — Formato de Documento Portátil (PDF), usado para documentos

---

### **Archivos de Python**
Los archivos de Python se reconocen por la extensión `.py`. Estos archivos contienen código escrito en el lenguaje de programación Python y son ejecutados por el intérprete de Python.

#### **¿Qué es un Archivo .py?**
Un archivo `.py` es un archivo de texto que contiene código Python. Puede incluir comandos, funciones o incluso clases. Estos archivos se usan para construir programas de Python que pueden realizar una variedad de tareas, desde aritmética simple hasta análisis de datos complejos.

---

### **Lo que Verás en el Árbol de Archivos**

1. **`first`**: Este es un archivo sin extensión. El sistema no sabe cómo manejarlo.
2. **`first.py`**: Este archivo tiene una extensión `.py`, indicando que es un script de Python. Codio lo reconocerá y aplicará resaltado de sintaxis para hacer el código más fácil de leer.

---

### **Abrir Archivos**
- **`first.py`**: Aparece con texto coloreado en el editor, que resalta palabras clave, cadenas y otras partes del código Python.
- **`first`**: Se abre como texto plano porque carece de una extensión que le indique al sistema operativo que el contenido es texto simple.
- **`first.txt`**: Se abre como texto plano porque tiene la extensión `.txt`, indicando que es un archivo de texto.

### **Cómo Ejecutar un Script de Python en Cuatro Pasos Simples**
1. Escribe tu código Python en un archivo con extensión `.py` (ej., `tuscript.py`).
2. Abre la **Terminal**.
3. Navega al directorio que contiene el archivo de Python.
4. Escribe el comando:  
```bash
   python3 nombredetuscript.py
```
   Luego presiona **Enter**.
