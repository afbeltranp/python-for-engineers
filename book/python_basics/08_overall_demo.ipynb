# Demostración General

## **Pregunta de Demostración General**

En esta demostración, trabajarás a través de una serie de tareas que involucran variables, tipos de datos, conversión de tipos y entrada del usuario. Sigue las instrucciones a continuación para completar el ejercicio.

### **Instrucciones: (`overall.py`)**

0. **`overall.py`**: Crea un archivo `overall.py`.

1. **Declarar una variable**:  
   Usa la función `input()` para solicitar al usuario que ingrese su nombre y almacénalo en una variable llamada `user_name`.

2. **Almacenar un entero**:  
   Usa la función `input()` para solicitar al usuario que ingrese su edad y almacénala en una variable llamada `user_age`. Convierte esta entrada a entero (`int()`) ya que se usará en una operación matemática.

3. **Capturar entrada del usuario**:  
   Usa la función `input()` para solicitar al usuario que ingrese su número favorito y almacénalo en una variable llamada `favorite_number`. Convierte esta entrada a entero (`int()`) ya que se usará en una operación matemática.
   
4. **Capturar más entrada del usuario**:  
   A continuación, usa la función `input()` nuevamente para solicitar al usuario que ingrese un número de punto flotante que represente el costo de algo que compró. Convierte esta entrada a flotante (`float()`) y almacénala en una variable llamada `item_cost`. (**no te preocupes por formatear a 2 decimales**)

5. **Realizar un cálculo**:  
   Suma el número favorito del usuario a su edad (almacenada en `user_age`). Almacena el resultado con una cadena formateada, por ejemplo: `"En 5 años, tu edad más tu número favorito será X."` (Se recomienda usar formato `f-string` para esto).

6. **Imprimir resultados**:  
   Imprime los valores de `user_name`, `user_age` y `favorite_number` en una oración. Luego, imprime el cálculo así como el costo del artículo. El formato esperado de estas impresiones se mostrará a continuación.

### **Ejemplo:**

**Entrada:**
```bash
Ingresa tu nombre: Alice
Ingresa tu edad: 25
Ingresa tu número favorito: 7
Ingresa el costo de tu artículo: 19.99
```

**Salida:**
```bash
¡Hola, Alice!
Tu edad es 25, y tu número favorito es 7.
En 5 años, tu edad más tu número favorito será 37.
El costo de tu artículo es 19.99 dólares.
```

**Pista**:
Para obtener la primera línea de la salida haz:
```bash
print(f"¡Hola, {user_name}!")
```
