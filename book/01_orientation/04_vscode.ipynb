# Casos de Prueba y VSCode

En VSCode estaremos usando *Pruebas de Código Estándar*, como la que se muestra a continuación para probar tu código. La idea es que un calificador automático ejecutará ciertos casos de prueba en tu código para tener en cuenta diferentes situaciones y casos extremos.

Aquí tenemos un archivo de python de ejemplo con un calificador de ejemplo. Luego viene seguido de un caso de prueba que fallará. Después de presionarlos puedes ver que cuando son correctos obtendrás una marca de verificación verde y cuando no son correctos puedes ver un resultado esperado y un resultado real.

### Acerca de las Pruebas de Código Estándar

La mayoría de las tareas en este curso tendrán **exclusivamente** pruebas de código estándar. Verás cómo calificas antes de marcar como completo para enviar. Esto se transferirá directamente a tu calificación, ¡así que es importante saber cómo funciona!

---
### Ejemplo de Caso de Prueba:

#### Paso 1 - Respuesta incorrecta
- Copia el código a continuación en el panel de la derecha (el archivo `example.py`)
```python
  print("¡Así es como se verá un caso de prueba incorrecto!")
```
- Presiona el botón `Check It!` en la prueba de código para ejecutar tu código.
- ¡Ahora deberías ver las salidas esperadas y reales!
  - Ahora puedes revisar **show diff** y probar la herramienta **text compare**.

#### Paso 2 - Corrigiendo la respuesta incorrecta
